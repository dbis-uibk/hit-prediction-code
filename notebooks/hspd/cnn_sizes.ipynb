```python
import tensorflow as tf
```

```python
input_shape = (4, 128, 1200, 1)
x = tf.random.normal(input_shape)

display('block 1')
layer_size = (128, 4)
y = tf.keras.layers.Conv2D(32, layer_size, activation='relu', input_shape=input_shape[1:])(x)
display(y.shape)
y = tf.keras.layers.MaxPool2D(pool_size=(1, 4))(y)
display(y.shape)

display('block 2')
layer_size = (1, 4)
y = tf.keras.layers.Conv2D(32, layer_size, activation='relu')(y)
display(y.shape)
y = tf.keras.layers.MaxPool2D(pool_size=(1, 4))(y)
display(y.shape)

# display('block 3')
# layer_size = (1, 3)
# y = tf.keras.layers.Conv2D(32, layer_size, layer_size, activation='relu')(y)
# display(y.shape)
# y = tf.keras.layers.MaxPool2D(pool_size=(1, 3))(y)
# display(y.shape)

display('block 4')
layer_size = y.shape[1:3]
y = tf.keras.layers.Conv2D(512, layer_size, layer_size, activation='relu')(y)
display(y.shape)
# y = tf.keras.layers.MaxPool2D(pool_size=(1, 2))(y)
display(y.shape)

display('block 5')
layer_size = (1, 1)
y = tf.keras.layers.Conv2D(512, layer_size, layer_size, activation='relu')(y)
display(y.shape)
# y = tf.keras.layers.MaxPool2D(pool_size=(1, 2))(y)
display(y.shape)

display('block 6')
layer_size = (1, 1)
y = tf.keras.layers.Conv2D(1, layer_size, layer_size, activation='relu')(y)
display(y.shape)
# y = tf.keras.layers.MaxPool2D(pool_size=(1, 2))(y)
display(y.shape)

display('flatten')
y = tf.keras.layers.Flatten()(y)
display(y.shape)
```
